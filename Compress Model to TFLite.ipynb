## h5 to tflite

```
import tensorflow as tf

converter = tf.lite.TFLiteConverter.from_keras_model_file("/content/drive/My Drive/eye_blink.h5", input_shapes={'input_2':[None,224,224,3]})
tflite_model = converter.convert()
open("converted_model_mobile.tflite", "wb").write(tflite_model)
```
